```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
pd.set_option('display.max.column', None)
```

```python
fifa_data = pd.read_csv("data.csv")
```

# Remove unncessory columns like unnamed, NaN(if has much NaN)

```python
fifa_data.head()
```

```python
fifa_data.info()
```

```python
fifa_data.drop(['Unnamed: 0','Loaned From'], axis='columns', inplace=True)
```

```python
fifa_data.head()
```

# Analyze all player wages

```python
fifa_data['Converted Wages'] = fifa_data.Wage.str.replace('€|K','').astype(float).apply(lambda x : x * 1000)
```

```python
sns.distplot(fifa_data['Converted Wages'], kde=False)
```

```python
fifa_data['Converted Wages'].value_counts().head(20)
```

# Analyse which Club or National Team has the best-rated players

```python
teams = fifa_data.Nationality.unique()
```

```python
teams
```

```python
total_countries = len(teams)
total_countries
```

```python
best_rated_players = fifa_data[fifa_data.groupby('Nationality')['Overall'].apply(lambda x: x > 85)].head(10)
```

```python
best_rated_players
```

```python
sns.distplot(best_rated_players.Overall)
```

# Note: Above observation is of top 10 Player's who has over all rating greater then 85

# Assess the strength of a team at a particular position

```python
strong_teams = fifa_data.groupby('Nationality')['Converted Wages'].sum().sort_values(ascending=False)
```

```python
strong_teams
```

```python
sns.distplot(strong_teams, kde= False)
```

# If sum of all wages of the particular country is high, then that means they are paying to quality players

# Analyse the team with the best dribbling speed

```python
avg = fifa_data.groupby('Club')['Dribbling'].mean().sort_values(ascending =False)
```

```python
len(avg)
```

```python
avg
```

```python
sns.distplot(avg, kde=False)
```

# Note: Juventus club has most dribbling speed
